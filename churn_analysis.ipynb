```python
df = pd.read_csv("bank-additional-full.csv", sep=';', engine='python', on_bad_lines='skip')
df.head()
```

```python
df.info()
df.columns
df.head()
```

```python
df['y'].value_counts(normalize=True)
```

```python
df['churn'] = df['y'].apply(lambda x: 1 if x == 'no' else 0)
```

```python
df.groupby('contact')['churn'].mean()
```

```python
df.groupby('month')['churn'].mean().sort_values()
```

```python
df.groupby('housing')['churn'].mean()
```

```python
df.groupby('loan')['churn'].mean()
```

```python
df.groupby('job')['churn'].mean().sort_values(ascending=False)
```

```python
total_customers = len(df)
conversions = (df['y'] == 'yes').sum()
non_conversions = (df['y'] == 'no').sum()

total_customers, conversions, non_conversions
```

```python
total_customers = len(df)
conversions = (df['y'] == 'yes').sum()
non_conversions = (df['y'] == 'no').sum()

value_per_customer = 1000

revenue_generated = conversions * value_per_customer
revenue_lost = non_conversions * value_per_customer

revenue_generated, revenue_lost
```

```python
print(f"Revenue Generated: ${revenue_generated:,.0f}")
print(f"Revenue Lost: ${revenue_lost:,.0f}")
```
